```python
# List of dataset names
datasets = ["dns", "ldap", "mssql", "netbios", "ntp", "snmp", "ssdp", "syn", "tftp", "udp", "udplag"]

# Dictionary to store loaded datasets
data = {}

# Load datasets and store them in the data dictionary
for dataset_name in datasets:
    file_path = f"C:\\Users\\user\\Documents\\jupyter\\Enhanced IDS\\Dataset\\{dataset_name}.csv"
    data[dataset_name] = pd.read_csv(file_path)

# Iterate through each dataset and check for missing values
for dataset_name, dataset in data.items():
    print(f"Missing Values - {dataset_name}")
    cnt = 0
    for col in dataset.columns:
        missing_mask = dataset[col].isnull() | (dataset[col] == '-')  # Check for NaN or '-'
        missing_count = missing_mask.sum()
        if missing_count > 0:
            cnt += 1
            print(f"\t{col}: {missing_count} missing values")
    print(f"Total {cnt} features with missing values\n")
```
